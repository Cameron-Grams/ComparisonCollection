```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# original source https://www.cdc.gov/nceh/lead/data/state/midata.htm

df = pd.read_excel('./MI_CountyLevelSummary_2014.xlsx', header=[2, 3])
```

```python
print(df.shape)
df.head()
```

```python
df.columns
```

```python
col = df.columns
columns_ = []
for i, c in enumerate(col):
    if i < 5:
        columns_.append(c[0])
    elif i < 9:
        columns_.append(c[0] + ': ' + c[1])
    else:
        columns_.append('Number: ' + c[1])
```

```python
# for c in columns_:
#     print(type(df[c][0]))
```

```python
df = df.values
df = pd.DataFrame(df, columns=columns_)
df = df.fillna(0)
df.iloc[:84,:]
```

```python
df.iloc[0,0]
```

```python
df.columns
```

```python
df_e = df[df['Children with Confirmed BLLs ≥ 5 µg/dL: Number'] > 0]
print(df_e.shape)
fig, ax = plt.subplots(1,1)
ax.hist(df_e['Children with Confirmed BLLs ≥ 5 µg/dL: Number'], bins=50);
plt.title("Blood levels above 5 micrograms/deciliter")
plt.ylabel('Number of counties')
plt.xlabel('Number of children');
```

```python
df_eh = df[df['Children with Confirmed BLLs ≥ 10 µg/dL: Number'] > 0]
print(df_eh.shape)
fig, ax = plt.subplots(1,1)
ax.hist(df_eh['Children with Confirmed BLLs ≥ 10 µg/dL: Number'], bins=50);
plt.title("Blood levels above 10 micrograms/deciliter")
plt.ylabel('Number of counties')
plt.xlabel('Number of children');
```
